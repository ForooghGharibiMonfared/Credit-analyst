```python
data_1 = pd.read_csv(file_path, skiprows=1)[:-2]
data_1.loc[:, columns].copy()
```

```python
#drop null column
data_1=data_1.dropna(axis='columns')

```

```python
#checking is there any nall left
null_values = data_1.isnull().values.any()

# Print the result
if null_values:
    print("There are null values in the DataFrame.")
else:
    print("There are no null values in the DataFrame.")
```

```python
#Checking how many classification group we have.
target_groups = data_1['loan_status'].value_counts()

# Print the count of unique groups
print("Number of unique groups in the target column:", len(target_groups))

# Print the names of the unique groups
print("Names of unique groups in the target column:")
print(target_groups.index.tolist())
```

```python
# we want to choose 4 classification group from our targate column
x_0 = {
    'Current': '0',
    'Fully Paid': '0',
    'In Grace Period': '1',
    'Late (16-30 days)': '2',
    'Late (31-120 days)': '2',
    'Charged Off':'3',
    'Issued': '3'
}

# Remove leading and trailing whitespaces from the 'loan_status' column
data_1['loan_status'] = data_1['loan_status'].str.strip()

# Replace values in the 'loan_status' column using the dictionary
data_1['loan_status'] = data_1['loan_status'].replace(x_0)

# Check the 'loan_status' column after replacement
print(data_1['loan_status'])
```

```python
#numerize our interest rate
data_1['int_rate'] = data_1['int_rate'].str.replace('%', '')
data_1['int_rate'] = data_1['int_rate'].astype('float') / 100

```

```python
#seperating of our features and targate value
targate_value=data_1['loan_status']
dataa= data_1.drop(columns=['loan_status' ])
# Select non-numeric columns
non_numeric_columns = dataa.select_dtypes(exclude=['number']).columns

# Convert non-numeric columns to boolean
dataa[non_numeric_columns] = dataa[non_numeric_columns].notnull()

# convert variable targate from non-numeric to numeric one since we need it for our NN structure
#since our classification group has an ordinally relation so we use Encoder labeling approaches.

from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
numTarg=encoder.fit_transform(targate_value)

```

```python
#features
dataa.shape[1]
```

```python
#sikit learn to split our data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(dataa,targate_value, random_state=1, stratify= targate_value)
x_train, x_vali, y_train, y_vali=train_test_split(x_train,y_train, test_size=0.2, random_state=1, stratify= y_train)
```

```python
#make the data compatible to tensorflow
import numpy as np
x_train = x_train.astype(np.float32)
y_train = y_train.astype(np.int32)
x_test = x_test.astype(np.float32)
y_test = y_test.astype(np.int32)
x_vali = x_vali.astype(np.float32)
y_vali = y_vali.astype(np.int32)
```

```python
import tensorflow as tf
X_train=tf.convert_to_tensor(x_train)
X_test=tf.convert_to_tensor(x_test)
Y_train=tf.convert_to_tensor(y_train)
Y_test=tf.convert_to_tensor(y_test)
x_vali=tf.convert_to_tensor(x_vali)
Y_vali=tf.convert_to_tensor(y_vali)
```

```python
#NN structure:
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=82, activation='relu', input_shape=(82,)),
    tf.keras.layers.Dense(units=42,  activation='relu'),
    tf.keras.layers.Dense(units=21,  activation='relu'),
    tf.keras.layers.Dense(units=4, activation='sigmoid')])
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'] )
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=82, epochs=14, validation_data=(x_vali, y_vali))
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_accuracy)
```
